```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
excel_file = '/Users/viveksasikumar/Downloads/movies.xls'
movies = pd.read_excel(excel_file)
```

```python
movies.head()
```

```python
movie_sheet1 = pd.read_excel(excel_file,sheetname=0,index_col=0)
movie_sheet2 = pd.read_excel(excel_file,sheetname=1,index_col=0)
movie_sheet3 = pd.read_excel(excel_file,sheetname=2,index_col=0)
```

```python
movies = pd.concat([movie_sheet1,movie_sheet2,movie_sheet3])
movies.head()
```

```python
movies.shape  #shape gives the DataFrame
movies.count().max()  #gives maximum number of rows
```

```python
sorted_by_gross = movies.sort_values(['Gross Earnings'], ascending = False)
sorted_by_gross['Gross Earnings'].head(10)
```

```python
sorted_by_gross['Gross Earnings'].head(20).plot(kind="bar",figsize=(15,10))
plt.show()
```

```python
movies.pivot_table('Gross Earnings',index=['Country'],  aggfunc='sum').sort_values('Gross Earnings',ascending=False).head(10).plot(kind='bar', figsize=(20,10))
```

```python
sorted_by_gross['Gross Earnings'].head(20).plot(kind='bar',figsize =(20,8))
```

```python
plt.matshow(movies.corr())
```

```python
import seaborn as sns
corr = movies.corr(method='spearman')
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,)
```

```python
select_att = movies.filter(['Duration','Budget','Imdb Score','Country','Director','Director','Facebook Likes - Actor 1','User Votes','Reviews by Crtiics'])
select_att.head()
corr1 = select_att.corr(method='pearson')
sns.heatmap(corr1, 
            xticklabels=corr1.columns.values,
            yticklabels=corr1.columns.values,)
```
